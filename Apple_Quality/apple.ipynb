```python
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
x=df[['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness', 'Ripeness', 'Acidity']]
y=df['Quality']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), ['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness', 'Ripeness', 'Acidity']),  # Apply MinMaxScaler to numerical features  # Apply OneHotEncoder to categorical features (if any)
    ],
)
```

```python
model = Pipeline([
    ("pre", preprocessor),
    ("log", LogisticRegression(penalty='l2',class_weight='balanced'))  
])
```

```python
model.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(y_test,pre)
```

```python
ma=confusion_matrix(y_test,pre)
```

```python
sns.heatmap(ma)
```
